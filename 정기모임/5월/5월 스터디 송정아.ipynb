```python
# 7) '객실등급' 컬럼과 '요금' 컬럼에 대한 데이터만 출력하시오.
df[['pclass','fare']]
```

```python
# 10) 객실 요금이 300 불 이상인 데이터에 대한 '성별', '나이', '객실등급', '생존여부' 컬럼을 출력하시오.

df[df['fare'] >= 300].loc[:, ['sex', 'age', 'pclass', 'survived']]
```

```python
# 11) '객실등급'의 데이터 값을 모두 소문자로 바꾸고 출력하시오. (First -> first, Second -> second, Third -> third)

df['class'].str.lower()
```

```python
# 12) '성인여부' 컬럼의 데이터 값 중 man, woman 을 adult 로 바꾸시오.

df.replace({'who': 'man'}, 'adult')
df.replace({'who': 'woman'}, 'adult')
```

```python
# 13) 데이터 셋의 각 컬럼에 대한 공백(결측치)의 갯수를 출력하시오.

df.isnull().sum()
```

```python
# 14) 데이터셋에는 총 몇 건의 데이터가 저장되어 있는지 출력하시오(전체 탑승자는 몇 명인가).

len(df)
```

```python
# 15) '생존여부' 컬럼의 데이터와 비교하여 '생존' 컬럼의 데이터 값이 동일한지를 확인하시오. ('생존' 컬럼의 데이터가 1 일때, '생존여부' 컬럼의 데이터가 no 이라면 서로 다른 데이터이다. 해당 데이터가 있는지를 찾으면 된다.)

df[(df['alive'] == 'no') & (df['survived'] == 1)]
```

```python
# 16) '생존' 컬럼과 동일한 값이 저장되어 있는 '생존여부' 컬럼을 삭제하시오
# 그리고 '등급' 컬럼과 동일한 값이 저장되어 있는 '객실등급' 컬럼을 삭제하시오.
df.drop(['survived', 'plcass'], axis=1)
```

```python
# 17) '탑승지코드' 컬럼과 '탑승지명' 컬럼 모두에서 데이터 값이 누락되어 있는 행(인덱스)을 찾으시오.

df[df['embarked'].isna() & df['embark_town'].isna()]
```

```python
# 18) '탑승지코드' 컬럼과 '탑승지명' 컬럼에서 결측치(데이터가 누락되어 있는)가 존재하는 행을 삭제하시오.
# (인덱스 61, 829 번 행에 결측치가 존재한다.)

df.dropna(subset=['embarked', 'embark_town'])
```

```python
# 19) 인덱스 59 ~ 63번 행에 대한 데이터만 출력하시오.

df.loc[59:63]
```

```python
# 20) '성인여부' 컬럼에서 child 데이터에 대응하는 '나이' 컬럼의 데이터가 누락되어 있는지 확인하시오.
# ('성인여부' 컬럼의 데이터가 child 일때, '나이' 컬럼의 데이터가 isnull 인 데이터를 찾으면 된다.)

df[(df['who'] == 'child') & (df['age'].isnull())]
```

```python
# 21) '나이' 컬럼의 데이터 값에 대한 평균을 출력하시오.

df['age'].mean()
```

```python
# 22) '나이' 컬럼의 결측치(빈 셀)를 '나이'에 대한 평균 값으로 채우시오.

df['age'].fillna(df['age'].mean())
```

```python
# 23) '선실코드' 컬럼의 데이터 값에는 어떤 종류의 값이 있는지 출력하시오(7 개).

df['deck'].unique()
```

```python
# 24) '요금' 컬럼에서 가장 높은 요금이 얼마인지 출력하시오.

df['fare'].max()
```

```python
# 25) 생존자에 대한 데이터만 출력하시오.
# ('생존' 컬럼의 데이터 값이 1 인 데이터를 추출하여 'survivor' 변수에 저장한 다음 출력하시오.)

df['survived'] == 1
```

```python
# 26) '객실등급' 별 승객수를 출력하시오.

df.groupby('class').count()
```
