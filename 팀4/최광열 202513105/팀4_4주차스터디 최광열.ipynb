while문으로 Hello World를 10번 출력하시오.
(조건문에는 숫자형 변수를 포함하시오)

```python
hello = 0
while hello<10:
  hello += 1
  print('hello World')
```

input()함수로 입력값을 받아 조건이 맞을 시 Hello World를 5번 출력하시오

```python
hello=int(input('입력된 숫자 '))
while hello>0:
  hello -= 1
  print('Hello World')
```

while True:로 시작하는 반복문을 Hello World 출력을 5번 반복 후 루프에서빠져나가도록 작성하시오.

```python
hello=5
while True:
  hello -= 1
  print('Hello World')
  if hello==0:
    break
```

while 반복문을 사용하여 1부터 10까지 숫자를 출력할 때 짝수는 출력하지 않고 시작점으로 돌아가도록 작성하시오.

```python
number = 0
while number<10:
  number += 1
  if number % 2 == 0:continue
  print(number)
```

for 문을 이용하여 구구단 2단을 출력하시오

```python
for i in range(1,10):
  for j in range(2,3):
    print(i,'*',j,'=',i*j)
```

for 문을 이용하여 ['바나나', '사과', '수박']리스트의 요소를 순서대로 하나씩 출력하시오.

```python
for fruit in ['바나나', '사과', '수박']:
  print(fruit)
```

for문과 range 함수를 이용하여 -7에서 -3까지 출력하시오(range함수에는 음수가 들어가도록 하시오)

for문과 range만을 이용하여 2부터 10까지 짝수만 출력하시오(반복문 내에서 연산 수행 x)

2개의 리스트를 입력받을 때 두 리스트의 길이가 같은지 불 자료형으로 반환하시오

```python
a=input('리스트 입력')
b=input('리스트 입력')
if len(a)==len(b):
  print('true')
else:
    print('false')
```

두개의 리스트를 입력받고 위에서 만든 함수를 이용해 두 리스트의 길이가 같다고 판별되면 두 리스트를 이용하여 딕셔너리를 만들어 반환하는 함수를 만드시오

```python
if len(a)==len(b):
  print(dict(zip(a,b)))
```

글로벌 변수를 이용하여 plus() 함수가 수행될 때마다 변수에 5가 더해지도록 함수를 정의하시오

```python
a=0

def plus():
  global a
  a += 5
  print(a)
```
